```python
from thinkdsp import play_wave, read_wave

# filename = '92002__jcveliz__violin-origional.wav'
filename = '263868__kevcio__amen-break-a-160-bpm.wav'

wave = read_wave(filename)
```

```python
from thinkdsp import decorate

# wave.plot()

start = 4.5
duration = 5.0
segment = wave.segment(start, duration)
segment_high = wave.segment(start, duration)
segment.plot()
decorate(xlabel='Time (s)')
```

```python
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

low

```python
spectrum.low_pass(5000)

spectrum.plot(high=10000)
decorate(xlabel='Frequency (Hz)')

filtered = spectrum.make_wave()
filtered.make_audio()
```

high

```python
spectrum = segment_high.make_spectrum()

spectrum.high_pass(5000)

spectrum.plot()
decorate(xlabel='Frequency (Hz)')

filtered = spectrum.make_wave()
filtered.make_audio()
```
