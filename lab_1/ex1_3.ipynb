```python
from thinkdsp import SinSignal, CosSignal
from thinkdsp import decorate
# Создаём два простых сигнала
sin_sig = SinSignal(freq=440, amp=1.0, offset=0)
# sin_sig1 = SinSignal(freq=880, amp=1, offset=0)
# sin_sig2 = SinSignal(freq=1320, amp=1, offset=0)
cos_sig = CosSignal(freq=880, amp=0.5, offset=0)

# Складываем их в сложный сигнал
signal = sin_sig + cos_sig#sin_sig1 + sin_sig2
signal.plot()
decorate(xlabel='Time (s)')
# Преобразуем сигнал в волну
wave = signal.make_wave(duration=1.0, start=0, framerate=11025)

# Прослушать сигнал
wave.apodize()
wave.make_audio()
```

```python
wave.plot()
```

```python
spectrum = wave.make_spectrum()

spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

Добавили частоты не кратные основной

```python
signal = (
    SinSignal(freq=440, amp=1.0) +
    CosSignal(freq=880, amp=0.5) +
    SinSignal(freq=1330.25, amp=0.3)
)

signal.plot() # Перестал быть переодическим
decorate(xlabel='Time (s)')
```

```python
wave = signal.make_wave(duration=1.0, framerate=11025)
wave.apodize()

spectrum = wave.make_spectrum()

spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```
